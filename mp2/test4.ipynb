```python
import matplotlib.pyplot as plt
import numpy as np

# Number of columns to plot
num_cols = len(red_wine.columns)

# Determine subplot grid size
ncols = 3
nrows = int(np.ceil(num_cols / ncols))

# Create figure and axes
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 5 * nrows))
axes = axes.flatten()  # Flatten in case it's 2D

# Loop through columns and plot
for i, column in enumerate(red_wine.columns):
    plt.sca(axes[i])  # Set current axis
    myplot.plot_normal_fit(red_wine[column])
    axes[i].set_title(column)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
