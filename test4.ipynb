```python
import pandas as pd

# Example DataFrame
data = {
    "Category": ["A", "B", "C", "D"],
    "Values": [10, 20, 15, 30]
}
df = pd.DataFrame(data)

print(df)
```

```python
# Make a column chart
df.plot(x="Category", y="Values", kind="bar")
```
