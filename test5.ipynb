```python
import f1_reader
import pandas as pd
```

```python
directory = "html"

results = {}
for year in range(2020, 2025):
    results[year] = f1_reader.read_f1_season_from_html_files_to_dictionary(year, directory)
```

```python
races_per_season = {year: len(season) for year, season in results.items()}
```

```python
print(races_per_season)
```

```python
my_series = pd.Series(races_per_season)
print(my_series)
my_series.plot(kind="bar")
```

```python
my_series.plot(kind="bar", title="Races per season", xlabel="Season", ylabel="Number of races")
```
